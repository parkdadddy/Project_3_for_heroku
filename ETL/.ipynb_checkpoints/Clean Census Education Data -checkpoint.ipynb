```python
edu_cols = ['SA2 Name', '2006',"2011", "2016"]
new_edu_df = edu_df[edu_cols].copy()
new_edu_df
```

```python
 new_edu_df = new_edu_df.groupby(['SA2 Name']).sum()
```

```python
new_edu_df = new_edu_df.rename(columns={"2006": "2006 Bachelor 25 and over",'2011':'2011 Bachelor 25 and over','2016':'2016 Bachelor 25 and over'})
new_edu_df
```

```python
total_df =  pd.read_csv('total_edu.csv')
total_df.set_index("Unnamed: 0")
```

```python
total_cols = ['SA2 Name','SA2 Code', '2006',"2011", "2016"]
total_df = total_df[total_cols].copy()
total_df
```

```python
total_df = total_df.rename(columns={"2006": "2006 Total",'2011':'2011 Total','2016':'2016 Total'})
total_df
```

```python
total_df = total_df.merge(new_edu_df, left_on='SA2 Name', right_on='SA2 Name')
```
