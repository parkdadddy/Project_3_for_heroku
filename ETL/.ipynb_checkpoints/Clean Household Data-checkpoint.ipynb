```python
import pandas as pd
import requests
import json
```

```python
household_df = pd.read_csv('Total Household 2006-2016.csv')
household_df.head()
```

```python
house_cols = ['SA2 Code', 'SA2 Name',"Household Size", "2006", "2011",'2016']
house_df= household_df[house_cols].copy()
```

```python
index = house_df[house_df['Household Size'] == 'Total'].index
```

```python
house_df.drop(index, inplace=True)
```

```python
house_df
house_df.dtypes
```

```python
house_df['Household Size'] = house_df['Household Size'].astype(int)
house_df.dtypes
```

```python
house_df["2006total"] = house_df["2006"] * house_df["Household Size"]
house_df["2011total"] = house_df["2011"] * house_df["Household Size"]
house_df["2016total"] = house_df["2016"] * house_df["Household Size"]
```

```python
house_df
```

```python
total_cols = ['SA2 Name', "2006total", "2011total",'2016total']
new_house_df= house_df[total_cols].copy()
```

```python
new_house_df = new_house_df.groupby(['SA2 Name']).sum()
new_house_df
```

```python
household_df_total = pd.read_csv('Total Household 2006-2016.csv')
household_df_total.head()
```

```python
house_cols = ['SA2 Code', 'SA2 Name',"Household Size", "2006", "2011",'2016']
house_df_new= household_df_total[house_cols].copy()
```

```python
house_df_new
```

```python
house_df_new = house_df_new.loc[house_df_new['Household Size'] == 'Total']
```

```python
household_df_new = pd.merge(house_df_new, new_house_df, how='inner', on = 'SA2 Name')
```

```python
household_df_new
```

```python
household_df_new["2006avg"] = household_df_new["2006total"] / household_df_new["2006"]
household_df_new["2011avg"] = household_df_new["2011total"] / household_df_new["2011"]
household_df_new["2016avg"] = household_df_new["2016total"] / household_df_new["2016"]
```

```python
household_df_new
```

```python
household_df_new["2006_2011change"] = household_df_new["2011avg"] - household_df_new["2006avg"]
household_df_new["2011_2016change"] = household_df_new["2016avg"] - household_df_new["2011avg"]
household_df_new["2006_2016change"] = household_df_new["2016avg"] - household_df_new["2006avg"]
```

```python
household_df_new
```

```python
household_df_new["2006_2011%"] = household_df_new["2006_2011change"] / household_df_new["2006avg"]
household_df_new["2011_2016%"] = household_df_new["2011_2016change"] / household_df_new["2011avg"]
household_df_new["2006_2016%"] = household_df_new["2006_2016change"] / household_df_new["2006avg"]
```

```python
household_df_new
```

```python
final_cols = ['SA2 Code', 'SA2 Name',"2006avg", "2011avg",'2016avg', "2006_2011%", "2011_2016%",'2006_2016%']
final_house_df = household_df_new[final_cols].copy()
```
