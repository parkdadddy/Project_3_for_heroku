```python
income_df_2006 = pd.read_csv('income2006_2011.csv')
income_df_2006.dtypes
```

```python
income_df_2006.head(60)
```

```python
income_2006 = {}
for sa2 in income_df_2006["SA2 Name"].unique():
    max = income_df_2006[income_df_2006["SA2 Name"]==sa2]['2006'].max()
    mode_income = income_df_2006[(income_df_2006["SA2 Name"]==sa2) &  (income_df_2006['2006']==max)]['Income'].values[0]
    income_2006[sa2] = mode_income
```

```python
new_income_df_2006 = pd.DataFrame(income_2006, index=[0])
new_income_df_2006
```

```python
income_2011 = {}
for sa2 in income_df_2006["SA2 Name"].unique():
    max = income_df_2006[income_df_2006["SA2 Name"]==sa2]['2011'].max()
    mode_income = income_df_2006[(income_df_2006["SA2 Name"]==sa2) &  (income_df_2006['2011']==max)]['Income'].values[0]
    income_2011[sa2] = mode_income
```
